```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = 'profiles.csv'
df = pd.read_csv(path)
traits = ["Extraversion", "Neuroticism", "Agreeableness", "Conscientiousness", "Openness"]

# ---- radar/spider plot ----
labels = traits
N = len(labels)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # close the loop

fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Start at the top and go clockwise
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# radial scale (adjust if needed)
vals = df[traits].to_numpy()
rmin = max(0, np.floor(vals.min() * 2) / 2)  # round down to 0.5
rmax = np.ceil(vals.max() * 2) / 2          # round up to 0.5
ax.set_ylim(rmin, rmax)

for _, row in df.iterrows():
    values = row[traits].tolist()
    values += values[:1]  # close the loop
    ax.plot(angles, values, linewidth=2, label=row["model"])
    ax.fill(angles, values, alpha=0.08)

ax.set_title("Big Five Features (Radar Chart)", pad=20)
ax.legend(loc="upper left", bbox_to_anchor=(1.05, 1.05))
plt.tight_layout()
plt.show()


```
